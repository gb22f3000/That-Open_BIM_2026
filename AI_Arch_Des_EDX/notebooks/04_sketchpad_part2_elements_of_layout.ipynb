# A Python-based Sketchpad - Part 2: Elements of a Layout

Create and modify geometrical elements of floor layouts: walls (lines), spaces (polygons), visualize with matplotlib, and structure layouts with dictionaries.

## Walls - Lines as Lists of Points

```python
x1, y1 = 0.2, 0.3
x2, y2 = 1.5, 2.7
print("Point 1 (2D): ", x1, y1)
print("Point 2 (2D): ", x2, y2)
```

```python
c1 = (x1, y1)
c2 = (x2, y2)
l = [c1, c2]
print("Line (2D): ", l)
```

## Spaces - Polygons

```python
c1 = (0, 0); c2 = (2.2, 0); c3 = (2.2, 1.5); c4 = (0, 1.5)
poly = [c1, c2, c3, c4]
print("Polygon (2D): ", poly)
```

## NumPy Arrays - Modifying Shapes

```python
import numpy as np
poly_array = np.array(poly)
print("Polygon as array:\n", poly_array)
```

```python
poly_array_t = poly_array + 2
print("Translated:\n", poly_array_t)
```

```python
poly_array_s = poly_array * 2
print("Upscaled:\n", poly_array_s)
poly_array_s = poly_array * 0.5
print("Downscaled:\n", poly_array_s)
```

```python
change = np.array([2, -1])
print("Independent translate:\n", poly_array + change)
print("Independent scale:\n", poly_array * change)
```

## Indexing and Slicing

```python
c2 = poly_array[1]
print("c2:\n", c2)
poly_array[1] = np.array([2.5, 0.35])
print("Modified polygon:\n", poly_array)
```

```python
c2_5 = np.array([[2.35, 0.95]])
poly_plus = np.concatenate((poly_array[:2], c2_5, poly_array[2:]))
print("With extra point:\n", poly_plus)
poly_min = np.concatenate((poly_array[:2], poly_array[3:]))
print("Without c3:\n", poly_min)
```

## Visualizing Shapes

```python
import matplotlib.pyplot as plt
poly_array = np.array(poly)
poly_closed = np.concatenate((poly_array, poly_array[:1]))
poly_x, poly_y = poly_closed[:, 0], poly_closed[:, 1]
plt.plot(poly_x, poly_y)
plt.show()
```

```python
poly1 = poly_closed + 2
poly2 = poly_closed * -3
poly3 = poly_closed * np.array([0.4, 1.5])
poly4 = np.concatenate((poly_closed[:1], np.array([[1, 0.5]]), poly_closed[1:3], np.array([[1, 1]]), poly_closed[3:])) * 6
plt.figure(figsize=(10, 10)); plt.axis('off'); plt.gca().set_aspect('equal')
for p in [poly_closed, poly1, poly2]:
    plt.plot(p[:,0], p[:,1])
plt.fill(poly3[:,0], poly3[:,1], color='purple')
plt.fill(poly4[:,0], poly4[:,1], color='yellow', alpha=0.25)
plt.show()
```

## Rectangle Function

```python
def get_rectangle(w, h, m=[0, 0]):
    """Returns a rectangle as a closed 2D numpy polygon."""
    m1, m2 = m[0], m[1]
    c1 = [m1 + w/2, m2 + h/2]
    c2 = [m1 - w/2, m2 + h/2]
    c3 = [m1 - w/2, m2 - h/2]
    c4 = [m1 + w/2, m2 - h/2]
    return np.array([c1, c2, c3, c4, c1])

def plot_polygon(ax, poly, **kwargs):
    ax.fill(poly[:, 0], poly[:, 1], **kwargs)

rect = get_rectangle(3, 4, [2, 2])
print(rect)
```

## Structuring a Layout with Dictionaries

```python
rect1 = get_rectangle(2, 2, [1, 2])
rect2 = get_rectangle(2, 2, [3, 2])
floor = {
    "room 1": [rect1, "red"],
    "room 2": [rect2, "blue"],
}
_, ax = plt.subplots(1, 1, figsize=(10, 10))
ax.axis('off'); ax.set_aspect('equal')
for name, (poly, color) in floor.items():
    plot_polygon(ax, poly, fc=color, ec='black', alpha=0.5)
plt.show()
```
