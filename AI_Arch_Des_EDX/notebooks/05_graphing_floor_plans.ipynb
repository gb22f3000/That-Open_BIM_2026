## Dataset - MSD

Download from [Kaggle: Modified Swiss Dwellings](https://www.kaggle.com/datasets/caspervanengelenburg/modified-swiss-dwellings).
Place `graph_out` pickle files in `../data/msd_graphs/`.

```python
data_dir = "../data/msd_graphs"
if not os.path.isdir(data_dir):
    print(f"WARNING: {data_dir} not found. Download MSD graph_out files first.")
else:
    file_list = os.listdir(data_dir)
    print(f"Found {len(file_list)} graph files")
```

```python
graphs = []
if os.path.isdir(data_dir):
    file_list = os.listdir(data_dir)
    for filename in file_list:
        if filename.endswith(".pickle"):
            G = load_pickle(os.path.join(data_dir, filename))
            graphs.append(G)
    print(f"Loaded {len(graphs)} graphs")
```

## Accessing Graph Data

```python
if graphs:
    G = graphs[9]
    print(G)
    print("Room types:", G.nodes("room_type"))
    print("First room geometry:", G.nodes("geometry")[0])
```

## Visualize Floor Plan Graphs

```python
if graphs:
    G = graphs[9]
    fs = 10
    fig, ax = plt.subplots(1, 1, figsize=(fs, fs))
    ax.axis("off"); ax.set_aspect("equal")
    plot_floor(G, ax, fs)
    plt.show()
```

```python
if len(graphs) >= 8:
    fs = 10
    fig, axs = plt.subplots(2, 4, figsize=(fs*4, fs*2))
    axs = axs.flatten()
    for i, ax in enumerate(axs):
        ax.axis("off"); ax.set_aspect("equal")
        plot_floor(graphs[i], ax, fs)
    plt.tight_layout(); plt.show()
```
